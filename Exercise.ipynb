What are the top 10 monitored products by State?

```python
df_2 = pd.read_excel('sample.xlsx')
df_2
#Load a new dataframe after get thet there are two states. "DISTRITO FEDERAL" AND "MÉXICO"
```

```python
df_3 = df_2[df_2['estado']=='DISTRITO FEDERAL']#CONDITIONAL STATEMENT FOR THIS STATE
distrito_fed_prods = df_3['producto'].value_counts()[:10]#GET THE 10 PRODUCTS PEOPLE IN DISTRITO FEDERAL BUYS
distrito_fed_prods#PRINT THE RESULTS OF THE VARIABLE
```

```python
x_axis = ('DETERGENTE P/ROPA ','LECHE ULTRAPASTEURIZADA','REFRESCO ','CHILES EN LATA','MAYONESA','SHAMPOO','TELEVISORES','JABON DE TOCADOR','GALLETAS DULCES','YOGHURT ')
plt.bar(x_axis,distrito_fed_prods)
```

```python
mexico_products = df_2[df_2['estado']=='MÉXICO']
y_df=mexico_products['producto'].value_counts()[:10]
```

```python
x_axis_df = ('REFRESCO ','DETERGENTE P/ROPA','SHAMPOO','CHILES EN LATA','TOALLA FEMENINA','MAYONESA','LECHE ULTRAPASTEURIZADA','TELEVISORES','YOGHURT ','DESODORANTE')
plt.bar(x_axis_df,y_df)
```

Which is the commercial chain with the highest number of monitored products?

i DECIDED HERE TO SHOW A PIE PLOT THAT REPRESENTS THE DISTRIBUTIONS OF THE COMMERCIAL CHAINS AS A WHOLE COMERCIAL MEXICANA IS THE BIGGETS COMERCIAL CHAIN WITH THE HIGHETS NUMBER OF MONITORED PRODUCTS 

```python
monitored_products = df_2.groupby('cadenaComercial')['producto'].count()
monitored_products
```

```python
plt.pie(monitored_products,
        explode=(0,0.1,0,0),
        labels=('ABASTECEDORA LUMEN','COMERCIAL MEXICANA','I.S.S.S.T.E.','WAL-MART'),
        autopct='%1.1f%%')
```
